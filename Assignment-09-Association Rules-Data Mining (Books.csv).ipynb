```python
# Import Liabraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Import dataset
book = pd.read_csv(r"C:\Users\Mohan Patil\OneDrive\Desktop\Datasets\book.csv")
book
```

```python
# Data preprocessing not required .As Data is already in transaction format.
```

# Apriori Algorithm

# 1. Association rules with 10% Support and 70% confidence

```python
# 10 % Support
rules_support=apriori(book,min_support=0.1,use_colnames=True)
rules_support
```

```python
# 70 % Confidence
rules=association_rules(rules_support,metric='lift',min_threshold=0.7)
rules
```

```python
## A leverage value of 0 indicates independence. Range will be [-1 1]
## A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```

```python
rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule
rules[rules.lift>1]
```

```python
# Visualization of obtained Rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
