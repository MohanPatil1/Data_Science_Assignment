```python
# import liabraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# import dataset
crime = pd.read_csv(r"C:\Users\Mohan Patil\OneDrive\Desktop\Datasets\crime_data.csv")
crime
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
from sklearn.preprocessing import StandardScaler
crime_norm=StandardScaler().fit_transform(crime)
crime_norm
```

## Dendrogram

```python
#Create Dendrograms
plt.figure(figsize=(10,7))
dendrograms=sch.dendrogram(sch.linkage(crime_norm,'single'))
```

```python
#Create Dendrograms
plt.figure(figsize=(10,7))
dendrograms=sch.dendrogram(sch.linkage(crime_norm,'complete'))
```

```python
#Create Dendrograms
plt.figure(figsize=(10,7))
dendrograms=sch.dendrogram(sch.linkage(crime_norm,'average'))
```

```python
#Create Dendrograms
plt.figure(figsize=(10,7))
dendrograms=sch.dendrogram(sch.linkage(crime_norm,'centroid'))
```

```python
# Create Hirarchical Clustering
h_clusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
h_clusters
```

```python
h = pd.DataFrame(h_clusters.fit_predict(crime_norm),columns=['cluster_id'])
h['cluster_id'].value_counts()
```

```python
# Add clusters to dataset
crime['cluster_id']=h_clusters.labels_
crime
```

```python
crime.groupby('cluster_id').agg(['mean']).reset_index()
```

```python
# Plot Clusters using scatter plot
plt.figure(figsize=(10,5))
plt.scatter(crime['cluster_id'],crime['Murder'],c=h_clusters.labels_)
```
