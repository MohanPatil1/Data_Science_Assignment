```python
# Import Liabraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# import dataset
airline=pd.read_csv(r"C:\Users\Mohan Patil\OneDrive\Desktop\Datasets\EastWestAirlines.csv")
airline
```

```python
airline2=airline.drop(['ID#'],axis=1)
airline2
```

```python
airline2.info()
```

# Standard Scaler function

```python
# Normalize heterogenous numerical data by using Standard Scaler
airline2_norm=StandardScaler().fit_transform(airline2)
airline2_norm
```

```python
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range(1,11):
    kmean=KMeans(n_clusters=i,random_state=2)
    kmean.fit(airline2_norm)
    wcss.append(kmean.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build Cluster algorithm using K=4

```python
# Cluster algorithm using K=4
cluster4=KMeans(4,random_state=30).fit(airline2_norm)
cluster4
```

```python
cluster4.labels_
```

```python
# Assign cluster to dataset
airline4=airline2.copy()
airline4['clusters4id']=cluster4.labels_
airline4
```

```python
# Compute the centroids for K=4 clusters with 11 variables
cluster4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
airline4.groupby('clusters4id').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10,7))
plt.scatter(airline4['clusters4id'],airline4['Balance'],c=cluster4.labels_)
```
